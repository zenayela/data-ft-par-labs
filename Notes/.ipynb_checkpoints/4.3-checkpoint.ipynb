```python
# number of cars per year
x=df.groupby('Year').Make.agg('count');

plt.plot(x,'bo--')
```

```python
def select_correct_df(df, type_of_fuel):
    d1=df.loc[df['Fuel Type']==type_of_fuel]
    d2=d1.groupby('Year')[['Combined MPG','City MPG']].mean()
    d3=d2.melt(var_name='MPG type',ignore_index=False)
    return d3
    
    
```

```python
for ix,value in enumerate(['Regular','Premium']):
    print(ix, value)
```

```python
fig, axes=plt.subplots(1,2, figsize=(12,6), sharey=True)
for ix,value in enumerate(['Regular','Premium']):
    sns.lineplot(data=select_correct_df(df,value),
                 y='value',
                 x='Year', 
                 hue='MPG type',
                ax=axes[ix])
    axes[ix].set_title(value+' type of fuel')
axes[1].get_legend().remove()
plt.show()
```

```python
fig, axes=plt.subplots(1,2, figsize=(12,6), sharey=True)
for ix,value in enumerate(['Regular','Premium']):
    d1=df.loc[df['Fuel Type']==value]
    d2=d1.groupby('Year')[['Combined MPG','City MPG']].mean()
    d3=d2.melt(var_name='MPG type',ignore_index=False)
    sns.lineplot(data=d3,
                 y='value',
                 x='Year', 
                 hue='MPG type',
                ax=axes[ix])
    axes[ix].set_title(value+' type of fuel')
axes[1].get_legend().remove()
plt.show()
```

```python
labels = ['Vegetarians', 'Non-Vegetarians', 'Vegans']
sizes = [95, 100, 20]
explode = (0.1, 0.1, 0.1)
fig1, ax1 = plt.subplots(figsize=(12,3))
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=0)
ax1.axis('equal')
plt.show()
```

```python
labels = ['Vegetarians', 'Non-Vegetarians', 'Vegans']
sizes = [40, 100, 20]
explode = (0, 0.1, 0)
fig1, ax1 = plt.subplots(figsize=(12,3))
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=340)
ax1.axis('equal')
plt.show()
```
